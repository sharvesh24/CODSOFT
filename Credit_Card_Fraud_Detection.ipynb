```python
df=df.drop(['Unnamed: 0', 'street','first', 'last','street', 'city', 'state', 'zip','dob','merchant'],axis=1)
```

```python
df['gender'] = df['gender'].map({'M': 0, 'F': 1})
```

```python
df
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
df = pd.concat([df, pd.DataFrame(ohe.fit_transform(df[['category']]).toarray(), columns=ohe.categories_[0])], axis=1)
df = df.drop(['category'], axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
df = pd.concat([df, pd.DataFrame(ohe.fit_transform(df[['job']]).toarray(), columns=ohe.get_feature_names_out())], axis=1)
df = df.drop(['job'], axis=1)
```

```python
df
```

```python
import calendar
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['day_of_week'] = df['trans_date_trans_time'].dt.dayofweek
df['day_of_week'] = df['day_of_week'].apply(lambda x: 'Weekday' if x<5 else 'Weekend')
df['hour_of_day'] = df['trans_date_trans_time'].dt.hour
df['hour_of_day'] = df['hour_of_day'].apply(lambda x: 'Morning' if x<12 else ('Afternoon' if x<17 else 'Evening'))
df['month'] = df['trans_date_trans_time'].dt.month
df['month'] = df['month'].apply(lambda x: calendar.month_name[x])
df['year'] = df['trans_date_trans_time'].dt.year
df['day_of_month'] = df['trans_date_trans_time'].dt.day
df['day_of_month'] = df['day_of_month'].apply(lambda x: 'First Half' if x<16 else 'Second Half')
```

```python
df
```

```python
df=df.drop('trans_date_trans_time',axis=1)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['day_of_week'] = le.fit_transform(df['day_of_week'])
df['hour_of_day'] = le.fit_transform(df['hour_of_day'])
df['month'] = le.fit_transform(df['month'])
df['day_of_month'] = le.fit_transform(df['day_of_month'])
df['trans_num'] = le.fit_transform(df['trans_num'])
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['is_fraud'], axis=1)
y = df['is_fraud']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train,y_train)
print(model.score(X_test,y_test))

y_pred=model.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion:")
print(conf_matrix)
```
