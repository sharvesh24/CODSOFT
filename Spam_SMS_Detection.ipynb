```python
import pandas as pd
```

```python
df=pd.read_csv("D:/Programs/CodSoft/Spam/spam.csv",encoding='latin1')
```

```python
df
```

```python
df=df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['v1'] = le.fit_transform(df['v1'])
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

vectorizer.fit(df['v2'])

feature_matrix = vectorizer.transform(df['v2'])
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
Target=df['v1']
X_train, X_test, y_train, y_test = train_test_split(feature_matrix,Target, test_size=0.2, random_state=42)
classifier = MultinomialNB()
classifier.fit(X_train, y_train)
predictions = classifier.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print(f'Accuracy: {accuracy}')
print(confusion_matrix(y_test,predictions))
```
